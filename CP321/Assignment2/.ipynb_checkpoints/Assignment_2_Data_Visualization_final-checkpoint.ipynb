## Question 1

Please use the same data set as in Assignment 1 for the following plot.

For the Regular Unleaded Gasoline price in "Toronto East/Est", visualize the time series for all timestamps. Smooth the time series using simple moving average.

```python
import pandas as pd
df = pd.read_csv('data/fueltypesall.csv',index_col='Date', parse_dates=True)
print(df.shape)
df
#"Regular Unleaded Gasoline"
df1= df[(df['Fuel Type'] == 'Regular Unleaded Gasoline')] 
df1 = df1.copy()
df1.drop(df1.columns.difference(['Toronto East/Est','Fuel Type']), 1, inplace=True)
print(df1.shape)

```

```python
#fig = plt.figure()
#plt.style.use("dark_background")
df1["Toronto East/Est"]= df1["Toronto East/Est"].rolling(27, center=True).mean() #smoothly out bi-annually
df1.plot()
plt.xlabel("Time(years)",fontsize='20')
plt.ylabel("Price in cents/litre", fontsize='20')
plt.grid(color="#DAF7A6", linestyle="--", linewidth=0.4)
plt.title("Trend of Regular Unleaded Gasoline price in Toronto East/Est over 30 years",fontsize='25')
```

## Question 2

The following web site provides the updated numbers for coronavirus. Please plot the number of new cases for each country on a world map. 

Note: (1) Please clearly specify the date when you collected the data, since the numbers are updated every day. (2) Ignore the locations that do not relate to a country (e.g. "Diamond Princess").

https://www.worldometers.info/coronavirus/

```python
import geopandas
import matplotlib.pyplot as pyp
pyp.style.use('ggplot')
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
data= pd.read_csv('data/question2data.csv')
dfinal = world.merge(data, how='left', left_on='name', right_on='Name')
dfinal2=dfinal.copy()
dfinal2[['New Cases']] = dfinal2[['New Cases']].fillna(value='0')
vmin, vmax = 0, 1500
sm = pyp.cm.ScalarMappable(cmap='Oranges', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
dfinal2[dfinal2['name'].isin(['Diamond Princess','Antarctica']) == False].plot(column='New Cases', cmap='Oranges',figsize=(20,15), linewidth=0.7)
plt.colorbar(sm, orientation="horizontal", fraction=0.030, pad=0.1, aspect = 30)
plt.title("New cases of Corona Virus in different countries(Mar 8,2020)",fontsize=25,color="black")
```
